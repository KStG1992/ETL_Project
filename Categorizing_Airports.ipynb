```python
import pandas as pd
```

```python
#read the csv files in
airport = "Resources/airports.csv"

#create df for airports
airport_df = pd.read_csv(airport)
```

```python
#view the data
airport_df.head()
```

```python
# Locating Airports in the US that have Small Airports
sm_arprts = airport_df.loc[(airport_df['iso_country']=='US') & (airport_df['type']=='small_airport')]
                                                                
# Dropping NaN in IATA                                                              
sm_arprts = sm_arprts.dropna(subset=['iata_code'])
```

```python
# Locating Airports in the US that have Medium Airports
md_arprts = airport_df.loc[(airport_df['iso_country']=='US') & (airport_df['type']=='medium_airport')]
                                                                
# Dropping NaN in IATA                                                              
md_arprts = md_arprts.dropna(subset=['iata_code'])
md_arprts
```

```python
# Locating Airports in the US that have Medium Airports
lg_arprts = airport_df.loc[(airport_df['iso_country']=='US') & (airport_df['type']=='large_airport')]
                                                                
# Dropping NaN in IATA                                                              
lg_arprts = lg_arprts.dropna(subset=['iata_code'])
lg_arprts
```

```python
# Creating CSV File
sm_arprts.to_csv('Resources/sm_arprts.csv')
md_arprts.to_csv('Resources/md_arprts.csv')
lg_arprts.to_csv('Resources/lg_arprts.csv')
```
